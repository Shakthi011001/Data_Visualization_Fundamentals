```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.plot([1,7,9,14])
```

```python
plt.plot([1,2,7,8],[1,9,7,14])
```

```python
#Plotting the graph just with red circles
plt.plot([1,2,7,8],[1,9,7,14], 'ro')
```

```python
#Plotting the graph just with green circles
a = np.array([[1,2,3,4],[5,6,7,8]])
plt.plot(a[0],a[1],'go')
```

```python
#Simple plotting,naming the axes and describing the style of the graph
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.plot(a[0],a[1],'b--')
```

```python
#Plotting two graphs with variation in their styles
t = np.linspace(0,5,20)
plt.plot(t,t**2,color='coral',linestyle='--')
plt.plot(t,4*t,color='indigo',linewidth=2)
```

```python
#Bar graphs for grade dataset
grades = {'Guru' : 70, 'Madhur':82,'Niyantha':92,'Harjit':76,'Lokesh':97}
plt.bar(range(len(grades)), list(grades.values()), color='indigo')
plt.xticks(range(len(grades)),list(grades.keys()))
plt.xlabel("Students")
plt.ylabel("Grades")
```

```python
#Histograms for 1000 random datas 
x=np.random.randn(1000)
plt.hist(x, color='darkturquoise')
```
