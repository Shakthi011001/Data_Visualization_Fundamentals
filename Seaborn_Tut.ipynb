```python
#importing libraries and we read the data and display it
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns

df = pd.read_csv('Pokemon.csv' , index_col=0 , encoding='unicode-escape')
df.head()
```

```python
#scatter plot 
sns.lmplot(x='Attack', y='Defense', data=df)
plt.show()
```

```python
#adding few arguments and here as Stage are sperated as different colors
sns.lmplot(x='Attack', y='Defense', data=df, fit_reg=False, hue='Stage')
plt.show()

```

```python
#box plot using df copy data frame
df_copy = df.drop(['Total' , 'Stage' , 'Legendary'], axis=1 )
sns.boxplot(data=df_copy)
```

```python
#Violing plot for df_copy data frame
sns.violinplot(data=df_copy)

plt.show()
```

```python
#Violin plot with naming x axes and y axes
plt.figure(figsize=(10,6))
sns.violinplot(x='Type 1', y='Attack' , data=df)
plt.show()
```

```python
#Dot heat map dispalying the correlation for the data frame 
corr = df_copy.corr()
sns.heatmap(corr)
```

```python
#histograms using seaborn

sns.distplot(df.Attack, color='blue')
```

```python
#count plots
#x ticks is used in order to see the names of the attributes clearly to prevent them from squishing
plt.xticks(rotation=-40)
sns.countplot(x='Type 1', data=df)
```

```python
#density plot using joint plot function
sns.jointplot(df.Attack, df.Defense, kind='kde', color='blue' )
```
